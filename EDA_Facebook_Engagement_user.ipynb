### Remove unnecessary columns.

```python
data=df.drop(['status','language','device_type'],axis=1)
```

```python
# df
```

### Count how many duplicate rows exist.

```python
data.duplicated().sum()
```

### Remove duplicate rows.

```python
data.drop_duplicates().inplace=True
data
```

### Handle missing values:

Fill numerical columns with mean/median.

Fill categorical columns with mode or “Unknown”.

```python
data['gender'].fillna(data['gender'].mode()[0],inplace=True)
data
```

```python
data['country'].replace(['###', ''], np.nan, inplace=True)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data
```

```python
data['country'] = data['country'].astype(str).str.strip().str.title()
data
```

```python
data['age'].fillna(data['age'].median(),inplace=True)
data
```

```python
data['age'].isnull().sum()
```

### Detect and correct typos in categorical columns (e.g., “male”, “Male”, “MALE”).

```python
data['gender'].replace(['male','Male','MALE'], 'M',inplace=True)

```

```python
data['gender'].replace(['female','Female','FEMALE'], 'F',inplace=True)
data
```
