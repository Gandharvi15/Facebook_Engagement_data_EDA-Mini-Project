### Remove duplicate rows.

```python
data.drop_duplicates().inplace=True
data
```

### Handle missing values:

Fill numerical columns with mean/median.

Fill categorical columns with mode or “Unknown”.

```python
data['gender'].fillna(data['gender'].mode()[0],inplace=True)
data
```

```python
data['country'].replace(['###', ''], np.nan, inplace=True)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data
```

```python
data['country'] = data['country'].astype(str).str.strip().str.title()
data
```

```python
data['age'].fillna(data['age'].median(),inplace=True)
data
```

```python
data['age'].isnull().sum()
```

### Detect and correct typos in categorical columns (e.g., “male”, “Male”, “MALE”).

```python
data['gender'].replace(['male','Male','MALE',' M','M '], 'M',inplace=True)

```

```python
data['gender'].replace(['x','other','Other',' x','X',' X'], 'Other',inplace=True)
```

```python
data['gender'].replace(['female','Female','FEMALE',' F'], 'F',inplace=True)
data
```

### Which are the top 5 countries with the most users?

```python
cou10=data['country'].value_counts().head(5)
cou10
```

```python


plt.figure(figsize=(10,5))
cou10.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 5 Countries with the Most Users')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

### What is the relationship between country and verification status?

```python
print(data['country'].nunique(),'unique entries')
print(data['is_verified'].unique())
```

```python
cv= data.groupby('country')['is_verified'].value_counts().unstack(fill_value=0)
cv.head(10)
```

```python
top_veri = cv.sort_values(by=True, ascending=False).head(10)
print(top_veri)
```

```python
top_veri.plot(kind='bar', stacked=True, figsize=(10,5), color=['#ff9999','#66b3ff'])
```

```python
data['is_verified'] = data['is_verified'].astype(bool)

veri_rate = (data.groupby('country')['is_verified'].mean().sort_values(ascending=False).head(10) * 100)
print(veri_rate)
```
